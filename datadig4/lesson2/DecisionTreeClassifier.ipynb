```python
# -*-encoding=utf-8-*-
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from pandas import Series, DataFrame

test_data = pd.read_csv("C:\Users\scc\PycharmProjects\untitled1\datadig4\lesson2\\train.csv")
train_data = pd.read_csv("C:\Users\scc\PycharmProjects\untitled1\datadig4\lesson2\\test.csv")
```
